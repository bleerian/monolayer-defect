```python
# Import statements
from matplotlib import pyplot as plt
from pymatgen.analysis.adsorption import *
from pymatgen.core import Lattice, Molecule, Structure
from pymatgen.core.surface import generate_all_slabs
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
```

```python
mos2_V_S = Structure.from_file("../test/MoS2_2x2_V_S.vasp")
slabs = generate_all_slabs(mos2_V_S, max_index=1, min_slab_size=8.0, min_vacuum_size=10.0)
mos2_001 = [slab for slab in slabs if slab.miller_index == (0, 0, 1)][0]

asf_mos2_001 = AdsorbateSiteFinder(mos2_001)
ads_sites = asf_mos2_001.find_adsorption_sites()
print(ads_sites)
assert len(ads_sites) == 4

fig = plt.figure()
ax = fig.add_subplot(111)
plot_slab(mos2_001, ax, adsorption_sites=True)
```
